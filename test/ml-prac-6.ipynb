10. Identify outliers for any two columns and take appropriate steps.

```python
Q1 = data.quantile(0.25) #for each column
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print("IQR",IQR)

lower=Q1-1.5 * IQR
print("------lOWER Range-----")
print(lower)

print("------Upper Range-----")
upper=Q3 + 1.5 * IQR
print(upper)

df_outliers = data[((data < lower) |(data > upper)).any(axis=1)]
print("outlier_shape: ",df_outliers.shape)
```

```python
df_out_final = data[~((data < lower) |(data > upper)).any(axis=1)] #removing outlier...remaining data
print(df_out_final.shape)
plt.boxplot(df_out_final.survived)
```

```python
Q1 = data["survived"].quantile(0.25)
Q3 = data["survived"].quantile(0.75)
IQR = Q3 - Q1
print("IQR: ",IQR)
lower=Q1-1.5 * IQR
print("lower",lower)
upper=Q3 + 1.5 * IQR
print("Upper",upper)
data.loc[(data["survived"] >upper),"survived"]=data["survived"].quantile(0.95)
data.loc[(data["survived"] <lower),"survived"]=data["survived"].quantile(0.05)   
```
