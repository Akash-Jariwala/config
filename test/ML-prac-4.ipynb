9. Create a custom dataframe and apply logical, query and filter methods on the same

```python
df = pd.DataFrame({'name':['Jane','John','Ashley','Mike','Emily','Jack','Catlin'],'ctg':['A','A','C','B','B','C','B'],'val':np.random.random(7).round(2),'val2':np.random.randint(1,10, size=7)})

df
```

```python
# Logical OR operator
df[df['val']>0.5]
```

```python
# logical > operator
df[df.name > 'Jane']
```

```python
# Multiple logical conditions
df[(df['val'] > 0.5) & (df.val2 == 1)]
```

```python
#Working with strings using isin method
names = ['John','Catlin','Mike']
df[df.name.isin(names)]
```

```python
#Usage of tilde sign as a NOT operator
df[df.name.str.startswith('J')] #name starts with 'j'
df[~df.name.str.startswith('J')] #name does not start with 'j'
```

```python
# Query method to apply conditions on therow/col
df.query('ctg == "B" and val > 0.5')
```

```python
# Display all the row/cols contains largest and smallest methods
df.nlargest(3, 'val') #returns 3 largest from 'val'
df.nsmallest(4, 'val2') #returns 4 largest from 'val2'
```

```python
# rows 3 and 4, all columns
df.iloc[3:5, :]
```

```python
df.index = ['a','b','c','d','e','f','g']
df.loc['b':'d',:]
```

10. Calculate the mean, median and mode for the following dataframe.

```python
dataMatrix = {"D1":[135, 137, 136, 138, 138],
"D2":[43, 42, 42, 42, 42],
"D3":[72, 73, 72, 72, 73],
"D4":[100, 102, 100, 103, 104] };

dataframe = pd.DataFrame(dataMatrix)
dataframe

#column wise mean, median, mode
means = [dataframe[m].mean() for m in dataframe.columns]
medians = [dataframe[m].median() for m in dataframe.columns]
modes = [dataframe[m].mode() for m in dataframe.columns]

print("Mean: {}\nMedian: {}\nMode: {}".format(means,medians,modes))
```
